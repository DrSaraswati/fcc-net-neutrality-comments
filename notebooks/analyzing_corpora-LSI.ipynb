```python
def parse_term(t):
    t = t.split('*')
    t[0] = float(t[0])
    t[1] = t[1].replace('"','')
    return tuple(t)

def parse_topic(topic):
    return [parse_term(t) for t in topic.split(' + ')]
```

```python
def fit_and_score(nclust, corpus_matrix):
    mbk = MiniBatchKMeans(init='random', n_clusters=nclust, n_init=10,
                          init_size=1000, batch_size=1000, verbose=True)
    mbk.fit(corpus_matrix)
    score = metrics.silhouette_score(corpus_matrix, mbk.labels_, sample_size=10000)
    del mbk
    return (nclust, score)

def fit_and_return(nclust, corpus_matrix):
    mbk = MiniBatchKMeans(init='random', n_clusters=nclust, n_init=10,
                          init_size=1000, batch_size=1000)
    mbk.fit(corpus_matrix)
    return mbk
        
def test_range_of_cluster_num(min_n, max_n, corpus_matrix):
    for nclust in xrange(min_n, max_n):
        nclust, score = fit_and_score(nclust, corpus_matrix)
        print(nclust, score)
        yield (nclust, score)
```

```python
results = [(n,s) for n,s in test_range_of_cluster_num(2, 11, tfidf_corpus_lsi_matrix.T)]
```

```python
results_df = pd.DataFrame(results)
results_df.columns = ["clusters", "silhouette"]
results_df
```

```python
def plot_topic_dist(corpus_df, topics=range(200), sharex=True, sharey=True):
    grp_num = 0
    
    grouped = corpus_df.groupby('cluster_prediction')
    
    fig, axes = plt.subplots(nrows=max(len(grouped), 2), ncols=1,
                             sharex=sharex, sharey=sharey)

    fig.set_figheight(5*len(grouped))
    fig.set_figwidth(20)
    
    for classlabel, group in grouped:
        means = group[topics].mean()
        top_topic = means.argmax()
        means.plot(ax=axes[grp_num],
                   title='class: ' + str(classlabel) + ' topic: ' + str(top_topic))
        grp_num += 1
        
plot_topic_dist(tfidf_corpus_lsi_df)
    
```

```python
topic_df = pd.DataFrame(topic_terms)
topic_df.columns = ['topic_id', 'weight', 'term']
```

```python
term_topic_counts = topic_df.term.value_counts()
term_topic_counts
```

```python
topic_spread = pd.DataFrame({'term': term_topic_counts.index.values,
                             'term_topic_count': term_topic_counts.values})
```

```python
topic_df = topic_df.merge(topic_spread, how='left', on='term')
```

```python
def plot_topic_dist_with_terms(corpus_df, top_terms, topics=range(200),
                               sharex=True, sharey=True):
    
    corpus_df.max()
    
    grp_num = 0
    
    grouped = corpus_df.groupby('cluster_prediction')
    
    fig, axes = plt.subplots(nrows=max(len(grouped), 2), ncols=1,
                             sharex=sharex, sharey=sharey)

    fig.set_figheight(5*len(grouped))
    fig.set_figwidth(20)
    
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    
    for classlabel, group in grouped:
        maxes = group[topics].median()
        top_topics = maxes.order(ascending=False).head(5).index.values
        top_values = maxes.order(ascending=False).head(5).values
        maxes.plot(ax=axes[grp_num],
                   title='class: ' + str(classlabel) + ' topics: ' + str(top_topics) + ' len: ' + str(len(group)))
        positions = [0.16 * n for n in range(1,6)]
        for i, top_topic in enumerate(top_topics):
            terms = top_terms[top_terms.topic_id == top_topic]['term'].tolist()
            textstr = '%s\n'* len(terms) % tuple(terms)
            textstr += 'max = %f' % (top_values[i])
            axes[grp_num].text(positions[i], 0.95, textstr, transform=axes[grp_num].transAxes,
                               fontsize=14, verticalalignment='top', bbox=props)
        grp_num += 1
        
plot_topic_dist_with_terms(tfidf_corpus_lsi_df, top_terms_per_topic)
```
