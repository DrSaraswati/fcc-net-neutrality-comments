```python
def fit_and_score(nclust, corpus_matrix):
    mbk = MiniBatchKMeans(init='random', n_clusters=nclust, n_init=10,
                          init_size=1000, batch_size=1000, verbose=True)
    mbk.fit(corpus_matrix)
    score = metrics.silhouette_score(corpus_matrix, mbk.labels_, sample_size=10000)
    del mbk
    return (nclust, score)

def fit_and_return(nclust, corpus_matrix):
    mbk = MiniBatchKMeans(init='random', n_clusters=nclust, n_init=10,
                          init_size=1000, batch_size=1000)
    mbk.fit(corpus_matrix)
    return mbk
        
def test_range_of_cluster_num(min_n, max_n, corpus_matrix):
    for nclust in xrange(min_n, max_n):
        nclust, score = fit_and_score(nclust, corpus_matrix)
        print(nclust, score)
        yield (nclust, score)
```

```python
results = [(n,s) for n,s in test_range_of_cluster_num(2, 11, corpus_lda_matrix.T)]
```

```python
results = sorted(results, key=lambda x: x[0])
```

```python
def plot_topic_dist(corpus_df, topics=range(100), sharex=True, sharey=True):
    grp_num = 0
    
    grouped = corpus_df.groupby('cluster_prediction')
    
    fig, axes = plt.subplots(nrows=max(len(grouped), 2), ncols=1,
                             sharex=sharex, sharey=sharey)

    fig.set_figheight(5*len(grouped))
    fig.set_figwidth(20)
    
    for classlabel, group in grouped:
        means = group[topics].mean()
        top_topic = means.argmax()
        means.plot(ax=axes[grp_num],
                   title='class: ' + str(classlabel) + ' topic: ' + str(top_topic))
        grp_num += 1
        
plot_topic_dist(corpus_lda_df)
    
```

```python
def parse_term(t):
    t = t.split('*')
    t[0] = float(t[0])
    return tuple(t)

def parse_topic(topic):
    return [parse_term(t) for t in topic.split(' + ')]
```

```python
topic_terms = []

for i in xrange(100):
    topic = lda_model.print_topic(i)
    topic_terms += [(i,weight,term) for weight,term in parse_topic(topic)]
```

```python
topic_df = pd.DataFrame(topic_terms)
topic_df.columns = ['topic_id', 'weight', 'term']
```

```python
topic_df
```

```python
term_topic_counts = topic_df.term.value_counts()
term_topic_counts
```

```python
topic_spread = pd.DataFrame({'term': term_topic_counts.index.values,
                             'term_topic_count': term_topic_counts.values})
```

```python
topic_df = topic_df.merge(topic_spread, how='left', on='term')
```

```python
topic_df
```

```python
np.max(topic_df.topic_id) + 1
```

```python
topic_df['inv_top_freq'] = np.log((np.max(topic_df.topic_id) + 1) / topic_df.term_topic_count)
#topic_df['inv_top_freq'] = (np.max(topic_df.topic_id) + 1) / topic_df.term_topic_count
#topic_df['inv_top_freq'] =  1 / topic_df.term_topic_count
topic_df['inv_top_freq']
```

```python
topic_df.inv_top_freq
```

```python
topic_df['adjusted_weight'] = topic_df.weight * topic_df.inv_top_freq
```

```python
top_five_terms_per_topic = topic_df.sort(
                                   ['topic_id', 'adjusted_weight'], ascending=False
                                  ).groupby(
                                   'topic_id'
                                  ).head(
                                    5
                                  )[['topic_id', 'term', 'adjusted_weight']]
```

```python
top_five_terms_per_topic
```

```python
def plot_topic_dist_with_terms(corpus_df, top_terms, topics=range(100),
                               sharex=True, sharey=True):
    grp_num = 0
    
    grouped = corpus_df.groupby('cluster_prediction')
    
    fig, axes = plt.subplots(nrows=max(len(grouped), 2), ncols=1,
                             sharex=sharex, sharey=sharey)

    fig.set_figheight(5*len(grouped))
    fig.set_figwidth(20)
    
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    
    for classlabel, group in grouped:
        means = group[topics].mean()
        top_topic = means.argmax()
        terms = top_terms[top_terms.topic_id == top_topic]['term'].tolist()
        textstr = '%s\n'* len(terms) % tuple(terms)
        means.plot(ax=axes[grp_num],
                   title='class: ' + str(classlabel) + ' topic: ' + str(top_topic))
        axes[grp_num].text(0.05, 0.95, textstr, transform=axes[grp_num].transAxes,
                           fontsize=14, verticalalignment='top', bbox=props)
        grp_num += 1
        
plot_topic_dist_with_terms(corpus_lda_df, top_five_terms_per_topic)
```
