## Cluster analysis

"Cluster analysis" is a general term for (often unsupervised) approaches to exploring a dataset by uncovering structure that's not immediately obvious or predefined. "Unsupervised" means we don't the _true_ grouping of our data points, and we're going to try to derive it from the data itself. Clustering is often done naively, with as few assumptions baked in to the methodology as possible. The goal is to let the data speak for itself, allowing data points to group together based on shared attributes. It's used in the field of data mining to get a sense of the "shape" of data. 

### Simple clustering example

A classic, concrete example of the use of clustering is the use of [k-means clustering](http://en.wikipedia.org/wiki/K-means_clustering) on [Fisher's Iris Dataset](http://en.wikipedia.org/wiki/Iris_flower_data_set). The data itself is the collected measurements of 50 samples from each of three species of Iris (a genus of flowers): setosa, virginica, and versicolor. By clustering each sample according to petal width, sepal length, and petal length, we can break up a bunch of data points into cohesive clusters. As it happens, these clusters correspond nicely to the true species groupings. For more on this example, check out the [tutorial on unsupervised learning](http://scikit-learn.org/stable/tutorial/statistical_inference/unsupervised_learning.html) at the scikit-learn site.

Again, K-means clustering is unsupervised. We're attempting to ask "What are the natural groupings of these data points," but we don't know what the "true" answer is. All we have to go on are: the shared attributes of our observations and the values of those attributes.

### Observations, features and models

Let's break down the Iris clustering example. Here's what happened.

 1. gather some flowers
 2. observe them
 3. cluster them according to the observations we made
 
*Step 1* is the world as-is: our raw observable world. In *Step 2*, we make observation, and turn raw reality into data: a set of records (one for each flower) where each record has a number of features (petal width, petal length, and sepal width), and each feature has a value (eg: 0.02, 1.4, 3.5).

Finally, in *Step 3*, we show that the values of those features can be used to discover a clustering **model**: spherical spaces whose centers and edges describe natural groupings of our observations. That's what we get from the k-means clustering algorithm. Choosing a clustering algorithm often depends on the values of the data itself: how they're distributed. In this example, k-means works very well, but it won't in every case, especially cases where natural groupings don't look like spheres with definite centers and roughly spherical boundaries.

It's worth mentioning, though, that *Step 2* also has a type of modeling going on: why did we measure the flowers in this way, and not others? There's a lot more to a flower than the three measurements we took into account. Why not measurements of other aspects of the flower (number of petals, radius of the stems)? Why not measurements of how they smell? Why not measurements of their color, weight, location, etc?

Having seen how successful this clustering was, it's easy to recognize that petal length, petal width and sepal length are useful observations for describing the natural groupings of these flowers. There are lots of other things that might have been observed, though, and added as features with their own values in our data model. Presumably the original investigators knew (or deduced) that those attributes held less information, and the choice was most likely informed by expert domain knowledge, but it's an illustration of a key point: **when clustering, it matters what you measured**. 

This is true of flower observations, where the number of reasonable features is at most in the dozens. You can fairly quickly try every possible combination of features and stop when you think you've got a good cluster model. However, in text clustering, the number of reasonable features can be in the hundreds of thousands or more. To create good cluster models, we're going to want to home in on which features are useful for clustering, and which aren't.

In machine learning, this is known as **feature selection**, and it's a common phase in any well-run study. Next time someone gives you data that they've clustered for you, ask them: which features did you use? It makes all the difference in the world.

### Interesting features of text data

When looking at a collection of textual observations (here, FCC comments), we'd really like to be able to cluster them into groups in a meaningful way. However, to get the advantages that come along with clustering, we'll have to decide what our features are: in other words, what measurements we'll be taking and what their values will represent. This can help us understand what it means to cluster texts using those features.

So what possible features are there for text? A few possibilities come to mind without too much effort:

- Length: Literally, how long is the text? This could be measured in words, characters, sentences, paragraphs, etc.
- Character Occurence: How many times does each type of character occur?
- Word Occurence: Which words occur in the text, and how many times?
- Topic Relevance: What areas of interest does the text pertain to?

If we measure the length of each document, we'd expect the clusters to be clusters of similarly sized documents. 

If we measure character frequency distributions, we'd have clusters that are characterized by shared distributions of characters (as it turns out, this is very useful for [language detection](http://odur.let.rug.nl/~vannoord/TextCat/textcat.pdf)).

In a project like this one, the measurement we're most interested in is topic relevance. If we measured each document for its topical content, we'd be able to form clusters characterized by shared topics. However, in this case (and in most cases), that's exactly the information we want and don't have.

The closest measurement we have is word occurrence, which should at least give us some idea of what's being covered in a given text. If we choose word occurrences as our measurements, we'll have clusters whose members share common words. If sharing common words can suffice as a proxy for sharing common topical content, then we should be fine.

Let's see what it takes to make word occurrence measurements.

### Normalizing and vectorizing text

The first step of doing large-scale textual analysis is representing the corpus (a fancy name for a collection of documents) in a memory-efficient way that is optimized for the kinds of analysis that you are interested in. Again, we're interested in word occurrence. An efficient way of recording word occurence is to *vectorize* text: give each unique word in the corpus a unique identifier and express each document as an array of those identifiers. So the sentences like these:

    'Remember when cable had no advertising? Just a schedule with movies all day and night, interrupted by short subjects and the like?'
    
Become expressed as a series of identifiers for the tokens (the units of language we care about) in them:

| type_id | type | token |
|---------|------|-------|
| 14 |	remember| Remember |
| 19 |	when | when |
| 5 |	cable | cable |
| 7 |	had | had |
| 13 |	no | no |
| 1	| advertising | advertising |

Terminology note: I'm using the word "type" intentionally. The "type" of each word is a bit of an abstraction over how the word originally appears in the text. Note that the token "Remember" shows up capitalized in the original text, but i've identifed its type as the lower-cased "remember." This helps when, for instance, counting all the occurrences of "rememeber" in the corpus, without having to worry about capitalization. It's not appropriate for all NLP tasks: capitals help for [named entity recognition](http://en.wikipedia.org/wiki/Named-entity_recognition) and [sentence boundary detection](http://en.wikipedia.org/wiki/Sentence_boundary_disambiguation), for instance.

Vectorizing in this way requires some bookkeeping (or not, if you use some [feature-hashing](http://en.wikipedia.org/wiki/Feature_hashing) techniques). A straightforward approach is to build a dictionary of all the tokens you might see, mapping them to unique identifiers, like so:

```python
s = """Remember when cable had no advertising?"""
tokens = [w.lower() for w in word_tokenize(s) if w.isalpha()]
types = set(tokens)
dictionary_lookup = {word_type: word_type_id for word_type_id, word_type in 
              enumerate(sorted(list(types)))}
dictionary_lookup
```

```python
sent_tokens = [w.lower() for w in word_tokenize("Remember when cable had no advertising?") if w.isalpha()]
pd.DataFrame(zip([dictionary_lookup[st] for st in sent_tokens], sent_tokens),
             columns=['type_id', 'type'])
```

When identifying the component parts of a sentence, it's typical practice to do a few things to normalize your text inputs:

 - get rid of non-text cruft like markup, page numbers, or punctuation
 - break up the text into tokens
 - move everything to a normalized version (ie, lowercase)
 - identify them with their associated types
 
That last step can be done in a few different ways, and how it's done is a choice that needs to be made according to the needs of a specific project. For this project, our mapping from token to type was taken care of by using [GPoSTTL](http://gposttl.sourceforge.net/), which is an implementation of the [Brill Tagger](http://en.wikipedia.org/wiki/Brill_tagger). Another alternative is Kyle Gorman's excellent [PPoSTTL](https://github.com/kylebgorman/pposttl), which is in Python and a bit easier to incorporate into other Python applications. We used GPoSTTL mostly because I already had some janky wrapper code lying around, but a more mature version of this codebase would probably use PPoSTTL.

Here's what the output of GPoSTTL looks like, to give an idea of what it's doing for us:

```python
!echo "Remember when cable had no advertising?" | gposttl
```

| type_id | type | token | POS |
|---------|------|-------|-----|
| 14 |	remember| Remember | VV |
| 19 |	when | when | WRB |
| 5 |	cable | cable | NN |
| 7 |	had | had | VHD |
| 13 |	no | no | DT |
| 1	| advertise | advertising | NN |
| 20 | ? | ? | SENT |

Okay, great, so now we have one word per line. We also have some other information, too: the second column is the [part of speech](https://www.englishclub.com/grammar/parts-of-speech_1.htm) of the word, and the third column is the **lemma**, which is basically the dictionary entry associated with each word. Lemmatization is an advanced form of [stemming](http://en.wikipedia.org/wiki/Stemming), where the variants of a word due to person, number, tense, etc are mapped to a shared form. Notice that _advertising_ gets changed to _advertise_. Similarly, _advertised_ and _advertises_ would be mapped to _advertise_. This lemma representation is what we'll be using as our type for each token. It will help us count instances of a word across its different forms.

The difference between lemmatization and the kind of stemming you'd get from, say, the [Porter Stemmer](http://tartarus.org/~martin/PorterStemmer/) is that the results are readable, displayable words most of the time. Porter would give us "advertis" for all of these forms, for instance. 

The part-of-speech tags used by GPoSTTL are based on the [Penn Tagset](http://www.comp.leeds.ac.uk/ccalas/tagsets/upenn.html), which is a standard reference in NLP software.

### Processing tons of text: quick-and-dirty parallelization

```python
%%bash
seq 1 10 | xargs -n1 -I {} echo "i am number {}"
```

```python
%%bash
for i in $(seq 1 10)
do
    echo "i am number ${i}";
done
```

Anyway! The real reason why xargs is great is that you can also use it to do some quick-and-easy parallel processing. The `-P` flag lets you run multiple processes at once when executing against a list of inputs. I put this to use in `scripts/process_all_json.sh`:

### Serialization

_relevant script:_ `scripts/serializing_corpus.py`

Another strength of the `gensim` package is the ability to serialize just about everything that you build with it: corpora, dictionaries, models, and even similarity interfaces (more on these later). Reading these objects from disk is MUCH more convenient than having to build them each time we want to do some work.

There are lots of options for serialization, including the native formats of other popular libraries for topic modeling, like [Mallet](http://mallet.cs.umass.edu/). There is also serialization support for [SVMLight](http://svmlight.joachims.org/) formatting, which is also recognized by the [LibSVM](http://www.csie.ntu.edu.tw/~cjlin/libsvm/)/[LIBLINEAR](http://www.csie.ntu.edu.tw/~cjlin/liblinear/)/[LibShortText](http://www.csie.ntu.edu.tw/~cjlin/libshorttext/) family of tools. We opted for the [Matrix Market](http://math.nist.gov/MatrixMarket/formats.html#MMformat) format, but that choice was mostly arbitrary. It's a good default, though, because it's particularly good for representing sparse matrices, and matrices don't get any sparser than the ones we're going to be generating.

Right about now, it'd be great to serialize bag-of-words representations produced by calling `my_dict.doc2bow()` on each document in our corpus. That way, we don't need to tokenize and identify tokens each time we read through the corpus. We'll do it once for everything, and remember the results.

Our corpus has now been stored as a sparse **document-term matrix**. The rows correspond to the documents, and the columns correspond to the words. A cell *(i,j)* in that matrix has value *c* where *c* is the number of times word *j* is found in document *i*.

For example, say the word "internet" is found 5 times in the 567th document. As we saw earlier, the dictionary assigned the id `10` to the term "internet." That means that cell (567, 10) will have the value 5. As we'll see, this is a really useful representation of a corpus.

### Corpus Transformations

_relevant script:_ `scripts/transform_corpus.py`

At this point, our corpus-as-matrix is ready to be modeled. With `gensim`, the general pattern for modeling is very similar to the pattern we used when creating a dictionary. It goes like this:

 - train a new model by iterating over every document in the corpus
 - use that model to transform your corpus
 - serialize **both** the model and the transformed corpus

For instance, one kind of useful transformation is to transform all of the values of the document-term matrix into values that are weighted according to each term's frequency in the corpus generally: the term frequency/inverse document frequency weights, or **[tf-idf](http://en.wikipedia.org/wiki/Tf%E2%80%93idf) weights**. These weights are so useful for text analysis and information retrieval tasks that their computation is often baked in to popular fulltext search engine software like Elasticsearch.

Clustering under k-means would work with this data, too. In contrast to the Iris data, we're dealing with 10 dimensions, but the math is the same. It's not possible to visually imagine 10 dimensions, but the same concept of a modeling natural groupings as a bunch of spheroid areas applies. 10-dimensional spheres are just like 3-dimensional spheres: there's a center, and all of the points of the sphere are equally distant from it. When it converges, k-means will find *k* spheres that are describe the natural grouping of the points. If we used these topics, we might expect to be able to set *k=2* and easily cluster documents into two groups that might represent the pro and anti sides of this debate. Even if that didn't work out, we'd almost certainly be able to consider clusterings at *k=3*, *k=4*, *k=5*, or more, and then be able to decide whether or not each cluster tends to be pro or anti.

The problem, though, is that we are counting two things that are unrealistic:

 - the ability to know the complete and true set of key concepts/topics
 - the ability to faithfully annotate millions of records by hand

# Tips

## setting params for hybrid clustering

### max branching
- greater values mean a flatter, more spread out tree

### min nodes
- smaller values mean more depth
- doesn't set the size of your smallest node: just the largest one at the deepest level (only one has to meet this requirement)

### max depth
- greater values means more aggressive recursion

###tradeoffs
- with higher branching
  - to control depth: increase jump threshold
  - to control splitting: increase min nodes
- with smaller min nodes
  - to control depth: higher jump threshold
  - to control splitting: less branching
- with greater max depth
  - to control splitting: increase jump threshold or min nodes
